```python
import nibabel as nib
import numpy as np
import os

import matplotlib.pyplot as plt
```

```python
orig_path = 'Delete/MRB_0169.nii_generated_volume.nii.gz'
orig_img = nib.load(orig_path)
orig_data = orig_img.get_fdata()
```

```python
import random
num_slices=5
slice_indices = random.sample(range(orig_data.shape[2]), num_slices)
fig, axes = plt.subplots(1, num_slices, figsize=(num_slices * 2, 4))

for i, idx in enumerate(slice_indices):
    axes[i].imshow(orig_data[:, :, idx], cmap='gray')
    axes[i].axis('off')
    axes[i].set_title(f'Slice {idx}')

plt.tight_layout()
plt.show()
```
